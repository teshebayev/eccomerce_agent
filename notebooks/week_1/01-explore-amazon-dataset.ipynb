## Filter items that have been observed for the first time in year 2022 or later

```python
def filter(data: dict) -> dict:
    filter =  False
    if int(data['details']['Date First Available'][-4:]) < 2022:
        filter = True
    return filter
```

```python
with open('../../data/meta_Electronics.jsonl', 'r') as fp:
    with open('../../data/meta_Electronics_filtered_2022_2023.jsonl', 'a', encoding='utf-8') as fp_out:
        with open('../../data/meta_Electronics_filtered_2022_2023_no_date.jsonl', 'a', encoding='utf-8') as fp_out_no_date:
            i = 0
            for line in fp:
                data = json.loads(line.strip())
                
                try:
                    filtered_data = filter(data)
                    if not filtered_data:
                        json.dump(data, fp_out)
                        fp_out.write('\n')
                        fp_out.flush()
                except:
                        json.dump(data, fp_out_no_date)
                        fp_out_no_date.write('\n')
                        fp_out_no_date.flush()
                i += 1
                if i % 1000 == 0:
                    print(f'Processed {i} lines')   
```

### split items into two categories: "has main category" , "does not have main category"

```python
def filter_category(data: dict) -> dict:
    filter =  False
    if data['main_category'] == None:
        filter = True
    return filter
```

```python
with open('../../data/meta_Electronics_filtered_2022_2023.jsonl', 'r') as fp:
    with open('../../data/meta_Electronics_filtered_2022_2023_with_category.jsonl', 'a', encoding='utf-8') as fp_out:
        with open('../../data/meta_Electronics_filtered_2022_2023_without_category.jsonl', 'a', encoding='utf-8') as fp_out_no_category:
            i = 0
            for line in fp:
                data = json.loads(line.strip())
                filtered_data = filter_category(data)
                if not filtered_data:
                        json.dump(data, fp_out)
                        fp_out.write('\n')
                        fp_out.flush()
                else:
                    json.dump(data, fp_out_no_category)
                    fp_out_no_category.write('\n')
                    fp_out_no_category.flush()
                i += 1
                if i % 10000 == 0:
                    print(f'Processed {i} lines')


```

#### Extract ratings that match sampled data.

```python
df_ratinds_100 = pd.read_json('../../data/meta_Electronics_filtered_2022_2023_with_category_ratings_100.jsonl', lines=True)
df_sample_1000 = pd.read_json('../../data/meta_Electronics_filtered_2022_2023_with_category_sample_1000.jsonl', lines=True)
```

```python
with open('../../data/meta_Electronics.jsonl', 'r') as fp:
    with open('../../data/meta_Electronics_filtered_2022_2023_with_category_ratings_100.jsonl', 'a', encoding='utf-8') as fp_out:
            id_list =set(df_ratinds_100['parent_asin'].values)
            i = 0
            for line in fp:
                data = json.loads(line.strip())
                if data['parent_asin'] in id_list:
                        json.dump(data, fp_out)
                        fp_out.write('\n')
                        fp_out.flush()
                i += 1
                if i % 1000 == 0:
                    print(f'Processed {i} lines')
```
